```python
#Observed that there was manual error inserting the entries,the duration was filled in the rating collumn
#so I made a function that take those entries checks if the word 'min' is present, if it is, it subistitute
#the NaN from duration to the value for rating and subistitutes the value for rating to Missing
def correct_entries(row):
    # Check for 'min' in the rating collumn and corrects it if so
    if ' min' in row['rating']:
        row['duration'] = row['rating']
        row['rating'] = 'N/A'
    return row

rows_to_correct = [5541, 5794, 5813]

df.loc[df.index.isin(rows_to_correct)] = df.loc[df.index.isin(rows_to_correct)].apply(correct_entries, axis=1)

df['duration'].isna().sum()
```

```python
#Created a variable that is equal to a group by of the added year and the added month with the .size()
#function and created a new index named count wich is the result of the .size() function
#Than pivot the data to be able to created a heatmap and finally plotted sad heatmap
monthly_additions = df.groupby(['added_year', 'added_month']).size().reset_index(name='count')

heatmap_data = monthly_additions.pivot(index='added_month', columns='added_year', values='count')

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt='g', linewidths=.5)
plt.title('Monthly Additions to Netflix Catalog Over Time')
plt.xlabel('Year')
plt.ylabel('Month')
plt.tight_layout()

plt.show()
```

```python
# Assuming 'df' is your DataFrame containing the data
# You may need to modify the column name accordingly
country_column = df['country']

# Remove leading and trailing whitespaces, convert to lowercase, and exclude 'unknown' entries
country_column = country_column.str.strip().str.lower()

# Filter out entries with 'unknown'
country_column = country_column[country_column != 'unknown']

# Split the entries in the 'country' column by commas
countries = country_column.str.split(', ')

# Flatten the list of lists
all_countries = [country for sublist in countries.dropna() for country in sublist]

# Count the occurrences of each country
country_counts = pd.Series(all_countries).value_counts()

# Select the top N countries (adjust N as needed)
top_countries = country_counts.head(10)

# Plotting the bar graph
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar')
plt.title('Top Countries by Frequency')
plt.xlabel('Countries')
plt.ylabel('Number of Movies added to Netflix')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
listed_in_column = df['listed_in']

# Split the entries in the 'listed_in' column by commas
categories = listed_in_column.str.split(', ')

# Flatten the list of lists and exclude 'International Movies'
all_categories = [category for sublist in categories for category in sublist if category != 'International Movies']

# Count the occurrences of each category
category_counts = pd.Series(all_categories).value_counts()

# Plotting the bar graph
plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar')
plt.title('Frequency of Categories')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```
